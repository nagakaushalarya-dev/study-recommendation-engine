```python
def generate_study_plan(subjects, hours):
    priority = {"weak": 3, "medium": 2, "strong": 1}
    plan = {}

    total_weight = sum(priority[level] for level in subjects.values())

    for subject, level in subjects.items():
        allocated_time = (priority[level] / total_weight) * hours
        plan[subject] = round(allocated_time, 2)

    return plan
```

```python
subjects = {
    "Maths": "weak",
    "Physics": "medium",
    "Chemistry": "strong"
}

hours_per_day = 4

study_plan = generate_study_plan(subjects, hours_per_day)

print("Personalized Study Plan:")
for subject, time in study_plan.items():
    print(f"{subject}: {time} hours/day")
```

```python
def advanced_study_plan(subjects, hours, exam_type):
    priority = {"weak": 3, "medium": 2, "strong": 1}
    exam_multiplier = {
        "Board Exam": 1.0,
        "Competitive Exam": 1.2
    }

    plan = {}
    insights = []

    total_weight = sum(priority[level] for level in subjects.values())
    multiplier = exam_multiplier.get(exam_type, 1.0)

    for subject, level in subjects.items():
        base_time = (priority[level] / total_weight) * hours
        adjusted_time = round(base_time * multiplier, 2)
        plan[subject] = adjusted_time

        if level == "weak":
            insights.append(f"Focus more on {subject}; revise basics and practice daily.")
        elif level == "medium":
            insights.append(f"Strengthen concepts in {subject} with mixed practice.")
        else:
            insights.append(f"Maintain performance in {subject} with periodic revision.")

    return plan, insights
```

```python
subjects = {
    "Maths": "weak",
    "Physics": "medium",
    "Chemistry": "strong"
}

hours_per_day = 4
exam_type = "Board Exam"

plan, insights = advanced_study_plan(subjects, hours_per_day, exam_type)

print("ðŸ“˜ Smart Study Plan\n")
for subject, time in plan.items():
    print(f"{subject}: {time} hours/day")

print("\nðŸ§  AI-Style Insights:")
for tip in insights:
    print("-", tip)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "difficulty_score": [3, 2, 1],  # weak=3, medium=2, strong=1
    "recommended_hours": [2.0, 1.3, 0.7]
}

df = pd.DataFrame(data)
```

```python
model = LinearRegression()
model.fit(df[["difficulty_score"]], df["recommended_hours"])
```

```python
def ml_recommendation(subjects):
    difficulty_map = {"weak": 3, "medium": 2, "strong": 1}
    ml_plan = {}

    for subject, level in subjects.items():
        score = difficulty_map[level]
        predicted_time = model.predict([[score]])[0]
        ml_plan[subject] = round(predicted_time, 2)

    return ml_plan

ml_plan = ml_recommendation(subjects)

print("ðŸ¤– ML-Based Study Recommendations:")
for subject, time in ml_plan.items():
    print(f"{subject}: {time} hours/day")
```

```python
import matplotlib.pyplot as plt

subjects_list = list(ml_plan.keys())
hours_list = list(ml_plan.values())

plt.figure(figsize=(6,4))
plt.bar(subjects_list, hours_list)
plt.title("ML-Based Study Time Allocation")
plt.xlabel("Subjects")
plt.ylabel("Hours per Day")
plt.show()
```
